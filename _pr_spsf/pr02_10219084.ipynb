Sebuah partikel bermuatan $q$ bergerak dengan kecepatan

$$ \vec{v}(t) = v_x(t)\hat{i} + v_y(t)\hat{j} $$

dalam ruang bermedan magnetik konstan

$$ \vec{B} = -k B_z $$

### a) Tuliskan hukum Newtonnya

Menurut hukum Newton 2,
$$
    \begin{equation} \tag{1}
    \begin{split}
        \sum{} \vec{F} = m \vec{a} 
    \end{split}
    \end{equation}
$$

karena hanya ada gaya oleh medan magnet, maka 
$$
    \begin{equation} \tag{2}
    \begin{split}
        \vec{F}_B = m \vec{a} \\
        q \vec{v} \times{} \vec{B} = m \frac{d\vec{v}}{dt} \\
    \end{split}
    \end{equation}
$$

Substitusikan $\vec{v}(t)$, maka
$$
    \begin{equation} \tag{3}
    \begin{split}
        q (v_x(t) \hat{i} + v_y(t) \hat{j}) \times{} (-\hat{k}B_z) = m \frac{ d\vec{v}(t) }{dt} \\
        q (v_x(t) B_z \hat{j} - v_y(t) B_z \hat{i}) = m \frac{ d\vec{v}(t) }{dt} \\
    \end{split}
    \end{equation}
$$
        

### b) Tuliskan persamaan diferensial terkopel antara kecepatan pada kedua arah

Persamaan 3 dapat dituliskan kembali sebagai
$$
    \begin{equation} \tag{4}
    \begin{split}
        q (v_x(t) B_z \hat{j} - v_y(t) B_z \hat{i}) = m \frac{d}{dt} (v_x(t)\hat{i} + v_y(t)\hat{j}) \\
    \end{split}
    \end{equation}
$$

Tinjau tiap komponen arah $\hat{i}$ dan $\hat{j}$, diperoleh

$$
    \begin{equation} \tag{5}
    \begin{split}
        - \frac{q B_z}{m} v_y(t) = \frac{d}{dt} v_x(t) \\
        \frac{q B_z}{m} v_x(t) = \frac{d}{dt} v_y(t) \\
    \end{split}
    \end{equation}
$$

### c) Selesaikan kedua persamaan diferensial sehingga dapat diperoleh $v_x(t)$, $v_y(t)$, $x(t)$, dan $y(t)$, lakukan secara teori

Tinjau Persamaan 5.1, lalu diferensialkan sehingga

$$
    \begin{equation} \tag{6}
    \begin{split}
        - \frac{q B_z}{m} \frac{d v_y(t)}{dt} = \frac{d^2 v_x(t)}{dt} \\
    \end{split}
    \end{equation}
$$

Substitusikan $\frac{d v_y(t)}{dt}$ dari Persamaan 5.2, maka

$$
    \begin{equation} \tag{7}
    \begin{split}
        - \frac{q B_z}{m} \frac{q B_z}{m} v_x(t) = \frac{d^2 v_x(t)}{dt} \\
        - \left( \frac{q B_z}{m} \right)^2 v_x(t) = \frac{d^2 v_x(t)}{dt} \\
    \end{split}
    \end{equation}
$$

Solusi persamaan tersebut adalah

$$
    \begin{equation} \tag{8}
    \begin{split}
        v_x(t) = k_1 sin\left( \frac{q B_z}{m}t \right) \
            + k_2 cos\left( \frac{q B_z}{m}t \right) \\
    \end{split}
    \end{equation}
$$

atau dapat dituliskan sebagai

$$
    \begin{equation} \tag{9}
    \begin{split}
        v_x(t) = A sin\left( \frac{q B_z}{m}t + \phi \right)
    \end{split}
    \end{equation}
$$

Sekarang tinjau Persamaan 5.2, lalu diferensialkan sehingga

$$
    \begin{equation} \tag{10}
    \begin{split}
        \frac{q B_z}{m} \frac{d v_x(t)}{dt} = \frac{d^2 v_y(t)}{dt} \\
    \end{split}
    \end{equation}
$$

Substitusikan $\frac{d v_x(t)}{dt}$ dari Persamaan 5.1, maka

$$
    \begin{equation} \tag{11}
    \begin{split}
        \frac{q B_z}{m} \left(-\frac{q B_z}{m} v_y(t)\right) = \frac{d^2 v_y(t)}{dt} \\
        - \left( \frac{q B_z}{m} \right)^2 v_y(t) = \frac{d^2 v_y(t)}{dt} \\
    \end{split}
    \end{equation}
$$

Solusi persamaan tersebut adalah

$$
    \begin{equation} \tag{12}
    \begin{split}
        v_y(t) = k_3 sin\left( \frac{q B_z}{m}t \right) \
            + k_4 cos\left( \frac{q B_z}{m}t \right) \\
    \end{split}
    \end{equation}
$$

atau dapat dituliskan sebagai

$$
    \begin{equation} \tag{13}
    \begin{split}
        v_y(t) = -A cos\left( \frac{q B_z}{m}t + \phi \right)
    \end{split}
    \end{equation}
$$

dengan $A$ adalah amplitudo dan $\phi$ adalah beda fase.

<br>
<br>

**Contoh kasus khusus**\
Diasumsikan $A=10$, $\phi=0$, $q=1$, $B_z=1$, $m=1$, maka Persamaan 9 dan Persamaan 13 menjadi
$$
    \begin{equation} \tag{14}
    \begin{split}
        v_x(t) = 10 sin(t) \\
        v_y(t) = -10 cos(t) \\
    \end{split}
    \end{equation}
$$

Berikut ini adalah hasil plot grafiknya

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
phi = 0
q = 1
Bz = 1
m = 1
```

```python
def vx(t):
    return A*np.sin(q*Bz/m*t + phi)
def vy(t):
    return -A*np.cos(q*Bz/m*t + phi)

arr_t = np.arange(0, 20, 1e-2)

plt.plot(arr_t, -vx(arr_t), arr_t, -vy(arr_t))
plt.title("Rotasi partikel pada medan magnet dengan penyelesaian secara teori")
plt.xlabel("Waktu")
plt.ylabel("Kecepatan")
plt.legend(["$v_x$", "$v_y$"])
plt.grid()

plt.show()
```

### d) Dapatkan solusi numeriknya

Lihat kembali persamaan diferensial terkopel pada Persamaan 5.

$$
    \begin{equation} \tag{5}
    \begin{split}
        - \frac{q B_z}{m} v_y(t) = \frac{d}{dt} v_x(t) \\
        \frac{q B_z}{m} v_x(t) = \frac{d}{dt} v_y(t) \\
    \end{split}
    \end{equation}
$$

Akan digunakan metode Runge-Kutta orde 4 untuk mendapatkan solusi numeriknya.

```python
# runge kutta 4th orde
def rk4(F,t,x,y,h):
    K0 = h*F(t,x,y)
    K1 = h*F(t + h/2.0, x + K0/2.0, y)
    K2 = h*F(t + h/2.0, x + K1/2.0, y)
    K3 = h*F(t +h, x + K2, y)
    return (K0 + 2.0*K1 + 2.0*K2 + K3)/6.0

# diferential equation
def vx_dot(t,vx,vy,q=1, Bz=1, m=1):
  return -q*Bz/m*vy
  
def vy_dot(t,vy,vx,q=1, Bz=1, m=1):
  return q*Bz/m*vx
```

```python
# initialization
arr_vx = [0]
arr_vy = [1]
arr_t = [0]
t_stop = 20
h = 1e-2

# looping for RK4 method
while arr_t[-1] < t_stop:
    h = min(h, t_stop-arr_t[-1])
    vx = arr_vx[-1] + rk4(vx_dot, arr_t[-1], arr_vx[-1], arr_vy[-1], h)
    vy = arr_vy[-1] + rk4(vy_dot, arr_t[-1], arr_vy[-1], arr_vx[-1], h)
    t = arr_t[-1] + h
    
    arr_t.append(t)
    arr_vx.append(vx)
    arr_vy.append(vy)

#plot kecepatan pada arah x dan y tiap waktu
plt.plot(arr_t, arr_vx, arr_t, arr_vy)
plt.title("Rotasi partikel pada medan magnet menggunakan metode Runge-Kutta ke - 4")
plt.xlabel("Waktu")
plt.ylabel("Kecepatan")
plt.legend(["$v_x$", "$v_y$"])
plt.grid()

plt.show()
```
