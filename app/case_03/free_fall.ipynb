# free-fall
simulate a free fall using finite difference method

## original example
url <https://pythonnumericalmethods.berkeley.edu/notebooks/chapter23.03-Finite-Difference-Method.html> [20221026]

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

n = 10
h = (5-0) / n

# Get A
A = np.zeros((n+1, n+1))
A[0, 0] = 1
A[n, n] = 1
for i in range(1, n):
    A[i, i-1] = 1
    A[i, i] = -2
    A[i, i+1] = 1

print(A)

# Get b
b = np.zeros(n+1)
b[1:-1] = -9.8*h**2
b[-1] = 50
print(b)

# solve the linear equations
y = np.linalg.solve(A, b)

t = np.linspace(0, 5, 11)

plt.figure(figsize=(10,8))
plt.plot(t, y)
plt.plot(5, 50, 'ro')
plt.xlabel('time (s)')
plt.ylabel('altitude (m)')
plt.show()
```

## modification
use the information at https://github.com/dudung/xeqn/blob/main/src/0020.md

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

n = 10
tbeg = 0
tend = 4
h = (tend-tbeg) / n

t1 = 3
y1 = 15

# Get A
A = np.zeros((n+1, n+1))
A[0, 0] = 1
A[n, n] = 1
for i in range(1, n):
    A[i, i-1] = 1
    A[i, i] = -2
    A[i, i+1] = 1

print(A)

# Get b
b = np.zeros(n+1)
b[0] = 0
b[1:-1] = -10*h**2
b[-1] = 0
print(b)

# solve the linear equations
y = np.linalg.solve(A, b)

t = np.linspace(tbeg, tend, n+1)

plt.figure(figsize=(10,8))
plt.plot(t, y)
plt.plot(t1, y1, 'ro')
plt.xlabel('time (s)')
plt.ylabel('altitude (m)')
plt.show()
```
